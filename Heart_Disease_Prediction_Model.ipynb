Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# loading csv data to a pandas dataframe
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
# print first five row of the dataset
heart_data.head()
```

```python
# print last five row of data set
heart_data.tail()
```

```python
# number of row and column in data base
heart_data.shape
```

```python
# getting more info about the data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measure about the data
heart_data.describe()
```

```python
# checking the distribution of target variable
heart_data['target'].value_counts()
```

1 --> Defective Heart
0 --> Healthy Heart

Splitting the feature and target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training logistic regression model
model.fit(X_train, Y_train)
```

Accuracy Score

Model Evaluation

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data :",training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data :",test_data_accuracy)
```

Building a Predictive System

```python
input_data = (56,0,0,200,288,1,0,133,1,4,0,2,3)

# change the input data to a numpy array
input_data_as_nummpy_array = np.asarray(input_data)

#reshape the numpy array as ae are predicting for only on instance
input_data_reshaped = input_data_as_nummpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
  print("The Person don't have heart Disease")
else:
  print("The Person Have Heart Disease")
```
